## Investigation Overview

This investigation looks at the categories of bike members that could be used to predict their ride durations in seconds. The main focus is on user type, member age range, and member gender utilizing univariate, bivariate, and multivariate visualizations.

## Dataset Overview

The data consists of ride durations and member cateogires of more than 453,159 rides in 2017. Some points were removed from the analysis due to inaccuracies or missing information. There are:
* 3670 bikes.
* 272 start stations.
* 272 end stations.
* 2 start/end hour categories: Day and Night.
* 7 start month categories: Jun 2017 to Dec 2017.
* 8 end month categories: Jun 2017 to Jan 2018.
* 2 user types: Customer and Subscriber.
* 4 age ranges: 18-30, 30-50, 50-70, and 70+.
* 3 gender types: Male, Female, and Other.

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe called trip_data_clean.csv
trip_data_clean = pd.read_csv('trip_data_clean.csv')
```

```python
# categorize member_age column into multiple segments in a new column called 'member_age_range'
bins_a = [18, 30, 50, 70, np.inf]
age_categories = ['18-30', '30-50', '50-70', '70+']
trip_data_clean['member_age_range'] = pd.cut(trip_data_clean['member_age'], bins_a, labels = age_categories)
```

## First Visualization

Question: How does the user type category segments compare in terms of count? Which segment is higher in terms of the amount?

```python
# count plot of start_hour_category
plt.figure(figsize=[14.70, 8.27])
sb.countplot(data = trip_data_clean, x = 'user_type', palette = "Blues");
plt.title('The User Type Category in Terms of Count', size = "15")
plt.xlabel('User Type');
plt.ylabel('Count');
```

The result of the plot shows that, in terms of user type, regular subscribers had a larger amount of rides than customers.

## Second Visualization

Question: How does the user type and member age range categories compare in terms of count? Which combination of segments is higher in terms of the amount?

```python
# count plot of user_type vs member_gender
plt.figure(figsize=[14.70, 8.27])
sb.countplot(data = trip_data_clean, x = 'user_type', hue = 'member_age_range', palette = 'Blues_r');
plt.title('The Relationship between the User Type and Member Age Range Categories in Terms of Count', fontsize = "15")
plt.xlabel('User Type');
plt.ylabel('Count');
```

The result of the plot, in terms of user type and member age range, shows that the amount of rides by subscribers in almost all age ranges is larger than the rides by customers. Subscribers of age ranges 18 to 30 and 30 to 50 had significantly more trips than older age ranges.

## Third Visualization

Question: What combination of the user type and member age range category segments result in the longest ride duration in seconds?

```python
# cat plot of user_type by duration_sec and member_age_range
fig, ax = plt.subplots(figsize=[14.70, 8.27])
sb.catplot(ax = ax, x = 'user_type', y = 'duration_sec', hue = 'member_age_range', palette = 'Blues_r', kind = 'bar', data = trip_data_clean);
ax.set(xlabel = 'User Type', ylabel = 'Ride Duration (sec)');
fig.suptitle('The Relationship between User Type and Member Age Range Categories and Ride Duration in Seconds', size = "15")
plt.close(2);
```

The result of the plot, in terms of user type & age range and duration, shows that longer trips took place by customers of all age ranges but specially customers who are 70+ due to possible variation.

## Fourth Visualization

Question: What combination of the user type and member gender category segments result in the longest ride duration in seconds?

```python
# cat plot of user_type by duration_sec and member_gender
fig, ax = plt.subplots(figsize=[14.70, 8.27])
sb.catplot(ax = ax, x = 'user_type', y = 'duration_sec', hue = 'member_gender', palette = 'Blues_r', kind = 'bar', data = trip_data_clean);
ax.set(xlabel = 'User Type', ylabel = 'Ride Duration (sec)');
fig.suptitle('The Relationship between User Type and Member Gender Categories and Ride Duration in Seconds', fontsize = "15")
plt.close(2);
```

The result of the cat plot, in terms of user type & gender and duration, shows that longer trips took place by customers of all genders specially females while other gender subscribers had the longer trips than male and female subscribers.
